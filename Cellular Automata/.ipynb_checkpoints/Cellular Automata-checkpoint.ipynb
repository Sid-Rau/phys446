```python
import numpy as np
import gzip as gz
import matplotlib.pyplot as plt
import random
```

```python
L = 100
# Initialize state
state = np.zeros((L,L))
state[0:L,0:(L//2)] = 1

# Number of steps/sweep
sweep = L**2
sweeps = 10000
recordSweeps = np.array([0,100,200,400,800,1600,3200,4800,6400,8000,9999])

# Record sizes of each state
state_sizes = np.zeros(sweeps)

def get_valid_neighbor(x, y, offset, xy):
    new_x = x + offset * xy
    new_y = y + offset * (1-xy)
    
    # Check if the new position is within bounds
    if 0 <= new_x < L and 0 <= new_y < L:
        return new_x, new_y
    return None

for i in range(sweeps):
    # Initialize random numbers for the sweep
    nx = np.random.randint(0, L, sweep)
    ny = np.random.randint(0, L, sweep)
    
    state_sizes[i] = sys.getsizeof(gz.compress(state.tobytes()))
    
    if i in recordSweeps:
        plt.imsave(f"sweep{i}.png", state)
        print("figure saved")
    
    for j in range(sweep):
        while True:
            x, y = nx[j], ny[j]
            offset = random.choice([-1, 1])
            xy = random.choice([0, 1])
            
            neighbor = get_valid_neighbor(x, y, offset, xy)
            
            if neighbor is not None:
                new_x, new_y = neighbor
                state[x,y], state[new_x,new_y] = state[new_x,new_y], state[x,y]
                break
            
            # If no valid neighbor found, pick a new random point
            nx[j] = np.random.randint(0, L)
            ny[j] = np.random.randint(0, L)

print("done")
```
